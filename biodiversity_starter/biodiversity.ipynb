### Analysis

#### Significance in Difference Between Between Species and Their Conservation Status?

```python
species_df.drop_duplicates(inplace  = True)
species_df.fillna('Not Protected', inplace = True)
sig_df = species_df[['category', 'conservation_status']]
def check(x):
    if x  == 'Not Protected':
        return x
    else:
        x = 'Protected'
        return x
def change(x):
    if x == True:
        x = 1
        return x
    else:
        x = 0
        return x
sig_df['conservation_status'] = sig_df['conservation_status'].apply(lambda x: check(x))
sig_df['Protected'] = sig_df['conservation_status'] != 'Not Protected'
sig_df['Not Protected'] = sig_df['conservation_status'] != 'Protected'
sig_df['Protected'] = sig_df['Protected'].apply(lambda x: change(x))
sig_df['Not Protected'] = sig_df['Not Protected'].apply(lambda x: change(x))
sig_df = sig_df[['category', 'Protected', 'Not Protected']]
sig_df = sig_df.groupby('category').sum()
print(sig_df)
```

Cecking the significance values using chi2 contingency:

```python
p_list  = {}
for i in range(len(sig_df.index) - 1):
    for j in range(i+1, len(sig_df.index)):
        chi2_table = sig_df.iloc[[i,j], [0,1]]
        x , p ,y,z = chi2_contingency(chi2_table)
        name1 = sig_df.index[i]
        name2 = sig_df.index[j]
        p_list[f'{name1},{name2}'] = round(p, 3)
```

Checking for any values below significance threshold:

```python
sig_values = []
for i,j in p_list.items():
      if j < 0.05:
            sig_values.append(i)
print('The total number of results is: ' + str(len(sig_values)))
print('They are:')
for i in sig_values:
    print(i)
```

Each of the species combinations listed above show signifcant differences in their conservation status rates. As you can see there 12 out of all the possible combinations that seem to indicate significant diffrences between the two. The most common type of animal in list is Nonvascular Plant.

#### Which Class of Animals Were Spotted the Most At Each Park?

```python
merged_df = pd.merge(species_df, obs_df, on = ['scientific_name'], how = 'inner')
sub_merged_df = merged_df[['category', 'park_name', 'observations']]
sub_merged_dfg = sub_merged_df.groupby(['park_name','category'])['observations'].sum().reset_index(name = 'total observations')
print(sub_merged_dfg)
sns.barplot(data = sub_merged_dfg, x = 'park_name', y = 'total observations', hue = 'category')
plt.xticks(rotation = 35)
plt.title('The Number of Observations Seen for Each Class in the Various National Parks')
plt.show()
plt.clf()
```

As we can see the vascular plants are the most observed class in all of the parks and cause the dimensions of bar chart to be blown out of proportion for the other classes. So knowing Vascular plants is the most observed we cna remove this from the data and observe how the other classes vary across the parks.

```python
sub_merged_dfg = sub_merged_dfg[sub_merged_dfg['category'] != 'Vascular Plant']
sns.barplot(data = sub_merged_dfg, x = 'park_name', y = 'total observations', hue = 'category')
plt.xticks(rotation = 35)
plt.title('The Number of Observations Seen for Each Class in the Various National Parks')
plt.show()
plt.clf()
```

The number of observations seen in each national park for each class follows a trend. With birds having the most observations, next being nonvascular plants, then mammals, then fish, then amphibians, then reptiles. This may indicate a trend of how numerous each of the classes are in national parks and could predict the proportions of each class in the overall ecosystem for national parks.

#### Most Observed Species in Each Park

```python
parks_dict = {}
names = merged_df['park_name'].value_counts().index
for i in names:
    parks_dict[i] = merged_df[merged_df['park_name'] == i]

np_max_dict = {}
for i,j in parks_dict.items():
    dummy = j['observations'].max()
    species = j['common_names'][j['observations'] == dummy].to_string(index = False)
    category = j['category'][j['observations'] == dummy].to_string(index = False)
    np_max_dict[i] = species
    print(f'In {i} the most observed species is a {category} which has the common names:\n {species}.')
```

As can be expected the most commonly observed species is a type of vascular plant in all national parks. However, each park has different plants which are the most common, most likely due to the different conditions observed at each park.
